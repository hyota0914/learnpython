```python
%matplotlib inline

import time
import poloniex
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pylab import rcParams
import seaborn as sns

# 横幅設定
rcParams['figure.figsize'] = 15, 10

# poloniex APIの準備
polo = poloniex.Poloniex()

# 5分間隔（サンプリング間隔300秒）で100日分読み込む
chart_data = polo.returnChartData('BTC_ETH', period=300, start=time.time()-polo.DAY*3, end=time.time())
df = pd.DataFrame(chart_data)

# 短期線：窓幅1日（5分×12×24）
data_s = pd.Series(df['close']).rolling(12*24).mean()
print(data_s)
# 長期線：窓幅5日（5分×12×24×5）
data_l = pd.Series(df['close']).rolling(12*24*5).mean()

# 一番簡単なプロット
plt.plot(df['close'], color='#7f8c8d')
plt.plot(data_s, color='#f1c40f')  # 短期線
plt.plot(data_l, color='#2980b9')  # 長期線
plt.show()
```
