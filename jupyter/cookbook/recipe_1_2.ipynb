```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import poloniex
import time
import datetime as dt
%matplotlib inline
%pylab inline
pylab.rcParams['figure.figsize'] = (15, 4)

polo = poloniex.Poloniex()
data = polo.returnChartData('USDT_BTC', period=300, start=time.time()-60*60*24*30, end=time.time())
df = pd.DataFrame.from_records(data)
df['date'] = df.apply(lambda x: dt.datetime.fromtimestamp(x['date']), axis=1)
df
```
